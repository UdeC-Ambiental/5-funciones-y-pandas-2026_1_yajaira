```python
import pandas as pd
```

```python
df = pd.read_csv("../data/resultados_experimentos.csv")
```

```python
df.columns
```

```python
def clasificar_biomasa(b):

    if b > 250:
        return "alta"

    elif b >= 150:
        return "media"

    else:
        return "baja"
```

```python
df["nivel_biomasa"] = df["biomasa"].apply(clasificar_biomasa)

df
```

```python
promedios = df.groupby("nivel_biomasa")[
    ["temperatura", "humedad", "luz"]
].mean()

promedios
```

```python
promedios.plot(kind="bar")
```
