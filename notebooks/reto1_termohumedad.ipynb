# Desarrollo del reto

Se desarrollaron funciones para resumir variables ambientales y clasificar las condiciones climáticas de cada sitio experimental.

```python
def resumen_termico(df):

    resumen = pd.DataFrame({
        "promedio_temperatura": [df["temperatura"].mean()],
        "max_temperatura": [df["temperatura"].max()],
        "min_temperatura": [df["temperatura"].min()],
        "promedio_humedad": [df["humedad"].mean()],
        "max_humedad": [df["humedad"].max()],
        "min_humedad": [df["humedad"].min()]
    })

    return resumen
```

```python
resumen = resumen_termico(df)
resumen
```

```python
def clasificar_clima(temp, hum):
    if temp >= 20 and hum >= 60:
        return "húmedo templado"
    elif temp >= 20 and hum < 60:
        return "seco templado"
    elif temp < 20 and hum >= 60:
        return "húmedo frío"
    else:
        return "seco frío"
```

```python
df["clima"] = df.apply(lambda row: clasificar_clima(row["temperatura"], row["humedad"]), axis=1)
df
```
