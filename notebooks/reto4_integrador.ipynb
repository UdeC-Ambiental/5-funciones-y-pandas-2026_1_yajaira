```python
import pandas as pd

df = pd.read_csv("../data/resultados_experimentos.csv")
df.head()
```

```python
df.head()
df.info()
df.describe()
df.columns
```

## Desarrollo del reto


```python
def calcular_promedio(columna):
    return columna.mean()

def valor_maximo(columna):
    return columna.max()

def valor_minimo(columna):
    return columna.min()
```

```python
print("Promedio temperatura:")
print(calcular_promedio(df["temperatura"]))

print("Temperatura máxima:")
print(valor_maximo(df["temperatura"]))

print("Temperatura mínima:")
print(valor_minimo(df["temperatura"]))
```

```python
import matplotlib.pyplot as plt

df["biomasa"].plot(kind="hist")

plt.title("Distribución de biomasa")
plt.xlabel("Biomasa")
plt.ylabel("Frecuencia")

plt.show()
```

## Interpretación de resultados

A partir de los análisis realizados se observó una relación entre las variables ambientales y la productividad del ecosistema representada en la biomasa.

Los resultados muestran que los niveles altos de biomasa se asocian con mayores valores de temperatura y luz, mientras que las categorías bajas presentan condiciones menos favorables para el desarrollo y productividad biológica.

El análisis por categorías permitió identificar tendencias importantes en el comportamiento de la biomasa según la disponibilidad de luz. En general, se evidencia que un aumento en la intensidad lumínica favorece la productividad, posiblemente debido al incremento de la actividad fotosintética.

Además, la clasificación mediante percentiles facilitó la comparación equilibrada de los datos, permitiendo agrupar las observaciones de manera más organizada y comprensible para el análisis estadístico.

En conjunto, los resultados indican que variables como la luz y la temperatura influyen significativamente en el comportamiento del ecosistema, demostrando la importancia del análisis de datos ambientales mediante herramientas como pandas para interpretar patrones ecológicos y apoyar futuros procesos de monitoreo ambiental.
